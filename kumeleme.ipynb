```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from sklearn.datasets import load_iris 
```

```python
iris=load_iris()
x=pd.DataFrame(iris.data)
x.columns=["sepal_length","sepal_width","petal_length","petal_width"]
```

```python
y= pd.DataFrame(iris.target)
y.columns=["target"]
```

```python
plt.figure(figsize = (14,7))
colormap = np.array(['red','blue','green'])
plt.subplot(1,2,1)
plt.scatter(x.sepal_length,x.sepal_width, c=colormap[y.target],
            s=40)
plt.title("sepal")
plt.subplot(1,2,2)
plt.scatter(x.petal_length,x.petal_width, c=colormap[y.target],
            s=40)
plt.title("petal")
plt.show()
```

```python
model=KMeans(n_clusters=3)
model.fit(x)
```

```python
plt.figure(figsize = (14,7))
colormap = np.array(['red','blue','green'])
plt.subplot(1,2,1)
plt.scatter(x.sepal_length,x.sepal_width, c=colormap[y.target],
            s=40)
plt.title("Gerçek Değerler")
plt.subplot(1,2,2)
plt.scatter(x.sepal_length,x.sepal_width, c=colormap[model.labels_],
            s=40)
plt.title("KMeans")
plt.show()
```

```python
print(model.labels_)
print(y.target)
```

```python
sm.accuracy_score(y.target,model.labels_)
```
